```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

C = 523.25
D = 587.33
E = 659.25
F = 698.46
G = 783.99
A = 880.00
Bb = 932.33
CC = 1046.50
```

```python
# Always with me

notes = [F, G, A, F, CC, A, G, CC, G, F, D, A, F, E ,\
         F, E, D, E, F, G, C, F, G, A, Bb, Bb, A, G, F, G]
       
durs = [0.25, 0.25, 0.25, 0.25, 0.75, 0.25, 0.50, 0.50, 0.50, 0.25, 0.25, 0.75, 0.25, 1.00 ,\
        0.25, 0.25, 0.50, 0.50, 0.25, 0.25, 0.50, 0.50, 0.25, 0.25, 0.50, 0.25, 0.25, 0.25, 0.25, 1.00]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th = phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
